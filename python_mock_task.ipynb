```python
import uuid

class Product:
  def __init__(self, name, description, price, category, discount):
    self.id = uuid.uuid4().hex
    self.name = name
    self.description = description
    self.price = price
    self.category = category
    self.discount = discount

products = []

```

```python
#add new product function

def add_product():
  print("Add new product : ")
  name = input("Enter Product Name : ")
  description = input("Enter Product Description : ")
  price = float(input("Enter Product Price : "))
  category = input("Enter Category Of Product : ")
  discount = float(input("Enter Product Discpunt in % : "))


  product = Product(name, description, price, category, discount)
  products.append(product)
  print("Product add successfully. Product ID : ", product.id)
```

```python
# Update product

def update_product():
  if not products:
    print("No products available to update. ")
    return

  print("Update product : ")
  product_id = input("Enter product id to Update : ")
  for product in products:
    if product.id == product_id:
      print("Product Found. Please provide Updated details : ")
      product.name = input("Enter new product Name : ")
      product.description = input("Enter new product Description : ")
      product.price = float(input("Enter new product Price : "))
      product.category = input("Enter new product Category : ")
      product.discount = float(input("Enter new product Discount in % : "))

      print("Product Updated Successfully.")

      return
  print("Product Not Found..")

```
